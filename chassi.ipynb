```python
##       W
##    /------/
##    ---|---  ---
##    |  |  |   |
##    |  |  |   | h
##    ---|---  ---
##       | Eixo

class Barra:
 
    # init method or constructor
    def __init__(self, w, h, E, ro, name):
        self.name = name
        self.cm_x = 0
        self.cm_y = 0
        self.cm_z = 0
        
        self.x = 0
        self.y = 0
        self.z = 0
        self.w = w
        self.h = h
        self.E = E
        self.ro = ro
        self.m = self.w*self.h*self.E*self.ro
        self.Iy = self.m*(self.w*self.w+self.E*self.E)
        self.Ix = self.m*(self.h*self.h+self.E*self.E)
        self.Iz = self.m*(self.w*self.w+self.h*self.h)
        self.log('BEGIN');
    
    def aproveitamento(self):
        print( 'Aproveitamento {} ---------------------------'.format(self.name) )
        print( 'x: {:.3} kgm²/kg'.format(self.Ix/self.m) )
        print( 'y: {:.3} kgm²/kg'.format(self.Iy/self.m) )
        print( 'z: {:.3} kgm²/kg'.format(self.Iz/self.m) )
    
    def log(self,TXT='LOG'):
        print( '{} {} ---------------------------'.format(TXT,self.name) )
        self.log_mass()
        self.log_I()
        self.log_CM()
 
    def log_mass(self):
        name = self.name
        print('> '+name + ' mass:')
        print('| mass[g]:', 1000*self.m)
        print(' ')
    
    def log_CM(self):
        name = self.name
        print('> '+name + ' CM:')
        print('| CM_x[mm]:', 1000*self.cm_x)
        print('| CM_y[mm]:', 1000*self.cm_y)
        print('| CM_z[mm]:', 1000*self.cm_z)
        print(' ')
    
    def log_I(self):
        name = self.name
        print('> '+name + ' I:')
        print('| Ix[kgm²]:', self.Ix)
        print('| Iy[kgm²]:', self.Iy)
        print('| Iz[kgm²]:', self.Iz)
        print(' ')
        return self.m
    
    def I(self):
        x = self.x
        y = self.y
        z = self.z
        self.Iy = self.m*(self.w*self.w+self.E*self.E + x*x + z*z )
        self.Ix = self.m*(self.h*self.h+self.E*self.E + y*y + z*z )
        self.Iz = self.m*(self.w*self.w+self.h*self.h + x*x + y*y )
    
    def move(self,x,y,z,LOG=True):
        self.x += x
        self.y += y
        self.z += z
        self.I()
        self.cm_x += x
        self.cm_y += y
        self.cm_z += z
        if(LOG): self.log('MOVE')
        
    
    def sub(self,sub_barra,LOG=True):
        self.Ix -= sub_barra.Ix
        self.Iy -= sub_barra.Iy
        self.Iz -= sub_barra.Iz
        
        self.cm_x = (self.cm_x*self.m - sub_barra.cm_x*sub_barra.m)/(-sub_barra.m+self.m)
        self.cm_y = (self.cm_y*self.m - sub_barra.cm_y*sub_barra.m)/(-sub_barra.m+self.m)
        self.cm_z = (self.cm_z*self.m - sub_barra.cm_z*sub_barra.m)/(-sub_barra.m+self.m)
        
        self.m -= sub_barra.m
        
        if(LOG): self.log('SUB')
    
    def add(self,sub_barra,LOG=True):
        self.Ix += sub_barra.Ix
        self.Iy += sub_barra.Iy
        self.Iz += sub_barra.Iz
        
        self.cm_x = (self.cm_x*self.m + sub_barra.cm_x*sub_barra.m)/(sub_barra.m+self.m)
        self.cm_y = (self.cm_y*self.m + sub_barra.cm_y*sub_barra.m)/(sub_barra.m+self.m)
        self.cm_z = (self.cm_z*self.m + sub_barra.cm_z*sub_barra.m)/(sub_barra.m+self.m)
        
        self.m += sub_barra.m
        
        if(LOG): self.log('ADD')
```

```python
Chassi = Barra(20e-2,15e-2,7e-3,2700,'chassi')

Janela1 = Barra(60e-3,120e-3,7e-3,2700,'janela 1')
Janela1.move(0,0,0,False)

Janela2 = Barra(45e-3,30e-3,7e-3,2700,'janela 2')
Janela2.move(-50e-3,50e-3,0,False)

Janela3 = Barra(45e-3,30e-3,7e-3,2700,'janela 3')
Janela3.move(50e-3,50e-3,0,False)

TOP = Barra(20e-2,15e-2,2e-3,2700,'TOP')
TOP.move(0,0,3.5e-3,False)

Chassi.sub(Janela1,False)
Chassi.sub(Janela2,False)
Chassi.sub(Janela3,False)
Chassi.add( TOP   ,False)

Chassi.log()
Chassi.aproveitamento()

# (tensão admissivel: 50MPa)
```
